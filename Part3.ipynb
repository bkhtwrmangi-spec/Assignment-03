Part 3: Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
# 1. Line Plot for Fare Trend of First 50 Passengers

plt.plot(df["Fare"][:50],color='green')
plt.title("Fare Trend of First 50 passengers")
plt.xlabel("Passenger Index")
plt.ylabel("Fare")
plt.grid(True)
plt.show()
```

```python
# 2. Bar Chart - Count of Passengers  by Class

class_counts = df["Pclass"].value_counts().sort_index()
plt.bar(class_counts.index, class_counts.values,color="red" )
plt.title("Passenger Count by Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# 3. Histogram 
plt.hist(df["Age"], bins=20,color="purple")
plt.title("age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# 4. Scatter 
plt.scatter(df["Age"], df["Fare"],color='orange')
plt.title("Fare vs Age of Passengers")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True)
plt.show()
```

```python
# Pie Chart 
survived = df["Survived"].astype(int).value_counts()

plt.pie(survived.values, labels=["Died", "Survived"], autopct="%1.1f%%")
plt.title("Survived Distribution")
plt.show()
```
