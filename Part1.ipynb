1: Load numerical columns from the dataset into Numpy arrays.

```python
import numpy as np

import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")

print(df)
```

```python
age = df["Age"].dropna().to_numpy()
fare = df["Fare"].dropna().to_numpy()
```

2: Show array properties: shape, dtype, ndim . 

```python
# Shape
print("Shape:", age.shape)

# Data Type
print("Data Type:", age.dtype)

# Number of Dimensions
print("Number of Dimensions:", age.ndim)
```

3: Perform vectorized operations, such as:
        . Mean, Median, std of any numeric column
        . Element-wise addition  / multipultion  

```python
print("Mean Age : " , np.mean(age))
print("Median Age :", np.median(age))
print("Std Age :", np.std(age))
```

```python
age_plus_2 = age + 2
print(age_plus_2)
```

```python
fare_double = fare * 2
print(fare_double)
```

4: Demonstrate:
          . Indexing
          . Boolean filtering
          . Slicing

```python
# indexing

print(age[0])   # first element
print(age[-1]) # last element
print(age[10:20])  # slice
```

```python
# Boolean Filtering

adult = age[age > 18]
print(adult)
```

```python
#  Slicing 

print(age[:10]) 
print(age[50:60])
print(age[-10:])
```

5: Use brodcasting to transform or scale value (e.g, normalize a column)

```python
fare_norm = (fare - fare.mean()) / fare.std()
print(fare_norm)
```

```python
age_scaled = age / age.max()
print(age_scaled)
```
