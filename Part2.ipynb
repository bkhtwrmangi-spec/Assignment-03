A . Data Loading & Cleaning
 
1. Load dataset into a DataFrame.

```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

2. Display basic info: head(), describe(), info().

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

3. Handle missing values (drop, fill, or impute ). 

```python
# Filling missing values

df["Age"]= df["Age"].fillna(df["Age"].median())
print(df["Age"])
```

```python
# Drop missing values

df= df.dropna()
print(df)
```

```python
# Fix data types

df["Survived"].astype("category")
df["Pclass"].astype("category")
print(df.dtypes)
print(df["Pclass"].dtype)
```

```python
pd.to_numeric(df["Fare"],errors='coerce')
```

```python
df.drop_duplicates()
```

B. Data Manipulation

1. Filtering rows based on conditions (e.g., age > 30).

```python
df[df["Age"] > 30]
```

2. Selecting columns & using .loc/ .iloc.

```python
df[["Age" , "Fare" , "Survived"]]
```

```python
df.loc[0:10, ["Age", "Fare", "Survived"]]
```

```python
df.iloc[0:5, 0:4]
```

3. Grouping & Aggregatons:
           . Mean, count, sum, min/mix of relevant columns.

```python
df.groupby("Pclass")["Age"].mean()
```

```python
df.groupby("Pclass")["PassengerId"].count()
```

```python
df.groupby("Pclass")["Age"].sum()
```

```python
df.groupby("Pclass")["Age"].min()
```

```python
df.groupby("Pclass")["Survived"].max()
```

4.  Merging/joining (if dataset has multiple CSVs. or creat a small second dataset and merge them).

```python
bonus_df = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Class_Name": ["Luxury", "Standard", "Economy"]
})
print(bonus_df)
```

```python
merged_df = pd.merge(df, bonus_df, on="Pclass", how="left")
merged_df.head()
```
